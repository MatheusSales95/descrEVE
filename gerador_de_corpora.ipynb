Gerador de Dados Fictícios (Dummy Data).

Script curto que combina sujeitos, verbos e adjetivos aleatoriamente para criar um arquivo com 300 frases (ou mais) automaticamente. Isso garante que você tenha variações suficientes para testar seu analisador.

```python
import random

# 1. Vocabulário para o Gerador (Baseado nas palavras que seu analisador conhece)
sujeitos = ["O produto", "O atendimento", "A entrega", "O manual", "O preço", "A qualidade", "O design", "A bateria", "O suporte", "O vendedor"]
verbos = ["é", "foi", "está", "parece", "chegou", "ficou"]
intensificadores = ["muito", "pouco", "extremamente", "realmente", "bastante", "super"]

# Adjetivos que seu PC.11 já sabe identificar
adjetivos_positivos = ["bom", "excelente", "ótimo", "rápido", "maravilhoso", "confiável"]
adjetivos_negativos = ["ruim", "péssimo", "horrível", "lento", "caro", "quebrado", "feio", "triste", "confuso"]
neutros = ["normal", "ok", "médio", "padrão", "azul", "vermelho"]

# 2. Função Geradora
def gerar_dataset_massivo(qtd_frases=5000):
    frases_geradas = []
    
    print(f"Gerando {qtd_frases} frases...")
    
    for _ in range(qtd_frases):
        # Sorteia um tipo de sentimento para esta frase (0=Neg, 1=Pos, 2=Neutro)
        tipo = random.choice(["pos", "neg", "neu"])
        
        suj = random.choice(sujeitos)
        verb = random.choice(verbos)
        inte = random.choice(intensificadores)
        
        if tipo == "pos":
            adj = random.choice(adjetivos_positivos)
            # Ex: "O produto é muito bom."
            frase = f"{suj} {verb} {inte} {adj}."
            
        elif tipo == "neg":
            adj = random.choice(adjetivos_negativos)
            # Ex: "A entrega chegou super lento."
            frase = f"{suj} {verb} {inte} {adj}."
            
        else:
            adj = random.choice(neutros)
            frase = f"{suj} {verb} {adj}."
            
        frases_geradas.append(frase)
        
    return "\n".join(frases_geradas)

# 3. Execução e Salvamento
conteudo_massivo = gerar_dataset_massivo(5000)

nome_arquivo_massivo = "dataset_reviews_random.txt"

with open(nome_arquivo_massivo, "w", encoding="utf-8") as f:
    f.write(conteudo_massivo)

print(f"✅ Arquivo '{nome_arquivo_massivo}' criado com sucesso com 5000 frases!")
print("Agora você pode apontar seu analisador para este arquivo.")
```

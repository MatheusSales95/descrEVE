Correção Ortográfica (Spell Checker)

O Conceito: O usuário digita errado ("casa" vira "caza"). O corretor busca em um corpus de referência qual palavra correta é mais "próxima" da errada. Essa proximidade é medida pela Distância de Edição (quantas letras preciso mudar para transformar A em B?).

```python
import difflib

# Nosso "Corpus" de palavras corretas conhecidas pelo sistema = vocabulario_correto
vocabulario_correto = ["pagamento", "boleto", "cartão", "cliente", "ciente","inconciente", "atraso", "crédito"]

palavra_errada = input("Digite a palavra: ") # Erro de digitação comum (troca de letras)

# A função get_close_matches busca as palavras mais parecidas
# n=1 quantidade de parentes proximos desejados
# cutoff=0.6 porcentagem de proximidade
sugestao = difflib.get_close_matches(palavra_errada, vocabulario_correto, n=5, cutoff=0.5)

print(f"Palavra digitada: {palavra_errada}")

if sugestao:
    print(f"Você quis dizer: {sugestao[0:]}?")
else:
    print("Nenhuma sugestão encontrada.")
```

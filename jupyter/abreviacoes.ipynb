O Conceito: Em redes sociais, "vc", "tb", "pq" são comuns. Para a IA, "vc" não existe no dicionário formal. Precisamos expandir isso antes de processar. O PDF sugere o uso de dicionários de mapeamento

```python
import re

glossario_internet = {
    "vc": "você",
    "tb": "também",
    "pq": "porque",
    "fds": "fim de semana",
    "abs": "abraços"
}

frase_usuario = "Oi vc vai estudar pln no fds? tb acho."
```

```python
def expandir_blindado(frase, dicionario):
    # PASSO 1: A Blindagem (Regex)
    # A lógica: Procure qualquer pontuação ([?.!,]) e substitua por " espaço + ela mesma + espaço "
    # O \1 refere-se ao que foi encontrado dentro dos parênteses
    frase_tratada = re.sub(r'([?.!,])', r' \1 ', frase)
    
    # Veja como ficou agora: "fds?" virou "fds ? "
    # print(f"DEBUG (Frase espaçada): {frase_tratada}") 
    
    # PASSO 2: Tokenização Segura (agora o split funciona)
    tokens = frase_tratada.split()
    
    tokens_corrigidos = []
    
    for token in tokens:
        # Busca no dicionário, se não achar, devolve o token original
        novo_token = dicionario.get(token.lower(), token)
        tokens_corrigidos.append(novo_token)
        # Teste para ver subs. das abreviações
        print(f'{tokens_corrigidos}')
    # PASSO 3: Reconstrução (Detokenization)
    frase_final = " ".join(tokens_corrigidos)
     
    return frase_final

print(f"Original:  {frase_usuario}")
print(f"Expandida: {expandir_blindado(frase_usuario, glossario_internet)}")
```

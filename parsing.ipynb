Parsing Sintático (Análise Profunda)

O Conceito: Diferente do POS Tagging (que diz se é Verbo ou Substantivo), o Parsing diz quem faz o quê. Ele identifica a estrutura da frase: Sujeito, Verbo, Objeto.


Exemplo: "O cliente pagou o boleto."

POS: cliente = Substantivo.

Parsing: cliente = Sujeito (nsubj) do verbo pagar.

```python
import spacy

try:
    nlp = spacy.load("pt_core_news_sm")
except:
    print("Modelo não carregado.")

# Frase passiva (mais difícil de analisar, mas o Parsing resolve!)
frase = "A decisão foi tomada pelo diretor."

doc = nlp(frase)

print(f"{'PALAVRA':<15} {'RELAÇÃO (DEP)':<15} {'CONECTADO A (HEAD)'}")
print("-" * 50)

for token in doc:
    # token.dep_ = Dependência (Relação sintática)
    # token.head.text = A palavra "pai" a qual este token está ligado
    print(f"{token.text:<15} {token.dep_:<15} {token.head.text}")

print("\n--- Análise Visual ---")
# Vamos achar o SUJEITO da frase (nsubj: passive nominal subject)
for token in doc:
    if "subj" in token.dep_:
        print(f"O Sujeito da oração é: {token.text}")
    if token.dep_ == "ROOT":
         print(f"O Verbo principal é: {token.text}")
```
